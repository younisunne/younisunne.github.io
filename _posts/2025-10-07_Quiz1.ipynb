```python
# 1. step 1
def my_perceptron(x, w, b):

  weighted_sum = sum(xi * wi for xi, wi in zip(x, w)) + b


  if weighted_sum >= 0:
    return 1
  else:
    return 0


x = [123, 456, 789]
w = [10, 11, 12]
b = -15000

print(my_perceptron(x, w, b))
```

```python
# 2. step 1

class MyClass:
    def __init__(self, value):
        self.my_attribute = value

    def my_method(self):
        return self.my_attribute + 1


class MyPerceptron:
    def __init__(self, w, d):
        self.my_w = w
        self.my_d = d

    def my_perceptron(self, x):

        weighted_sum = sum(xi * wi for xi, wi in zip(x, self.my_w)) + self.my_d


        if weighted_sum >= 0:
            return 1
        else:
            return 0

# step 2

x = [123, 456, 789]
w = [10, 11, 12]
d = -16000

my_instance = MyPerceptron(w, d)
result = my_instance.my_perceptron(x)
print(result)
```
